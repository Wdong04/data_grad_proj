```python
asteroids_data = asteroids_data.drop(["Orbiting Body", "Equinox","Name","Neo Reference ID","Orbit ID",
 "Epoch Date Close Approach","Close Approach Date","Orbit Determination Date","Est Dia in KM(max)",
 "Est Dia in M(min)","Est Dia in M(max)", "Est Dia in Miles(min)" ,"Est Dia in Miles(max)",
 "Est Dia in Feet(min)", "Est Dia in Feet(max)", "Relative Velocity km per hr","Miles per hour",
 "Miss Dist.(lunar)", "Miss Dist.(kilometers)", "Miss Dist.(miles)","Jupiter Tisserand Invariant", 
 "Epoch Osculation","Semi Major Axis", "Orbital Period"], axis = 1)
asteroids_data["Hazardous"] = asteroids_data["Hazardous"].astype("int")
asteroids_data = asteroids_data[asteroids_data["Orbit Uncertainity"] < 8]
```

```python
fig, axs = plt.subplots(nrows = 3, ncols=5, figsize = (50, 30))
axs = axs.flatten()
for i in range(15):
    sns.histplot(data=asteroids_data, x=asteroids_data.columns[i], hue="Hazardous", kde = True, ax=axs[i])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.model_selection import train_test_split, KFold, cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
```

```python
# logistic regression model
lr_model = LogisticRegression()
Y = asteroids_data['Hazardous']
X = asteroids_data.drop(columns='Hazardous')
X = (X-np.mean(X, axis=0))/np.std(X,axis=0)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=20)
lr_model.fit(X_train, Y_train)
Y_train_predict = lr_model.predict(X_train)
Y_test_predict = lr_model.predict(X_test)
cm_lr = confusion_matrix(Y_test, Y_test_predict)

#result
print ("Confusion Matrix : \n", cm_lr)
print('Accuracy (train): %.3f' % accuracy_score(Y_train, Y_train_predict))
print('Accuracy (test): %.3f' % accuracy_score(Y_test, Y_test_predict))
print('Precision (train): %.3f' % precision_score(Y_test, Y_test_predict))
print('Precision (test): %.3f' % precision_score(Y_test, Y_test_predict))
print('Recall (train): %.3f' % recall_score(Y_test, Y_test_predict))
print('Recall (test): %.3f' % recall_score(Y_test, Y_test_predict))
print('AUC: %.3f' % roc_auc_score(Y_test, Y_test_predict))
```

```python
# decision_tree_model
# process dataset
x = asteroids_data.drop(['Hazardous'],axis=1)
y = asteroids_data['Hazardous']
x, y = asteroids_data.iloc[: , :-1], asteroids_data.iloc[: , -1]
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=20 , test_size=0.3)

# build & train decision_tree_model model
decision_tree_model = tree.DecisionTreeClassifier(random_state = 20)
decision_tree_model = decision_tree_model.fit(x_train, y_train)

# predict
decision_tree_model.predict(x_test)

# accuracy
yhat_train = decision_tree_model.predict(x_train)
yhat_test = decision_tree_model.predict(x_test)
cm_dtc = confusion_matrix(y_test, yhat_test)

print ("Confusion Matrix : \n", cm_dtc)
print('Accuracy (train): %.3f' % accuracy_score(y_train, yhat_train))
print('Accuracy (test): %.3f' % accuracy_score(y_test, yhat_test))
print('Precision (train): %.3f' % precision_score(y_test, yhat_test))
print('Precision (test): %.3f' % precision_score(y_test, yhat_test))
print('Recall (train): %.3f' % recall_score(y_test, yhat_test))
print('Recall (test): %.3f' % recall_score(y_test, yhat_test))
print('AUC: %.3f' % roc_auc_score(y_test, yhat_test))

# Get feature importances
importances = decision_tree_model.feature_importances_
importances = pd.DataFrame([x_train.columns, importances]).T
importances.columns = ['Feature', 'Importance']
importances = importances.sort_values('Importance')

# Bar chart
fig, ax = plt.subplots(1, figsize=(10, 6))
plt.barh(importances['Feature'], importances['Importance'])
ax.set_xlabel('Gini Importance of Random Forest')
ax.set_title('Feature Importances')
plt.show()

# # decision tree
plt.figure(figsize=(15,15))
tree.plot_tree(decision_tree_model, feature_names = asteroids_data.columns,
              class_names = ["hazardous", "non-hazardous"],
              rounded = True, filled = True, impurity=False, fontsize=10)
plt.show()

# Tune hyperparameters: max_depth, min_samples_leaf
model = tree.DecisionTreeClassifier(random_state=0)
cv = KFold(n_splits=10, shuffle=True, random_state=0)
params = {'max_depth': np.arange(2, 15), 'min_samples_leaf': np.arange(1, 6)}
cv_model = GridSearchCV(model, param_grid=params, scoring='accuracy', refit=True, return_train_score=True, cv=cv)
cv_model.fit(x_train, y_train)
cv_results = pd.DataFrame(cv_model.cv_results_)

rfc_best = cv_model.best_estimator_
rfc_best.fit(x_train, y_train)
yhat_train = rfc_best.predict(x_train)
yhat_test = rfc_best.predict(x_test)

print('Best maximum depth: %i' % cv_model.best_params_['max_depth'])
print('Best minimum samples leaf: %i' % cv_model.best_params_['min_samples_leaf'])
print('CV accuracy: %.3f' % cv_model.best_score_)

print('Accuracy (train): %.3f' % accuracy_score(y_train, yhat_train))
print('Accuracy (test): %.3f' % accuracy_score(y_test, yhat_test))
print('Precision (train): %.3f' % precision_score(y_test, yhat_test))
print('Precision (test): %.3f' % precision_score(y_test, yhat_test))
print('Recall (train): %.3f' % recall_score(y_test, yhat_test))
print('Recall (test): %.3f' % recall_score(y_test, yhat_test))
print('AUC: %.3f' % roc_auc_score(y_test, yhat_test))
```

```python
# random forest
rfc = RandomForestClassifier(random_state=0)
rfc.fit(x_train, y_train)
yhat_train = rfc.predict(x_train)
yhat_test = rfc.predict(x_test)

cm_rfc = confusion_matrix(y_test, yhat_test)
print ("Confusion Matrix : \n", cm_rfc)
print('Accuracy (train): %.3f' % accuracy_score(y_train, yhat_train))
print('Accuracy (test): %.3f' % accuracy_score(y_test, yhat_test))
print('Precision (train): %.3f' % precision_score(y_test, yhat_test))
print('Precision (test): %.3f' % precision_score(y_test, yhat_test))
print('Recall (train): %.3f' % recall_score(y_test, yhat_test))
print('Recall (test): %.3f' % recall_score(y_test, yhat_test))
print('AUC: %.3f' % roc_auc_score(y_test, yhat_test))

# Get feature importances
importances = rfc.feature_importances_
importances = pd.DataFrame([x_train.columns, importances]).T
importances.columns = ['Feature', 'Importance']
importances = importances.sort_values('Importance')

# Bar chart
fig, ax = plt.subplots(1, figsize=(10, 6))
plt.barh(importances['Feature'], importances['Importance'])
ax.set_xlabel('Gini Importance of Random Forest')
ax.set_title('Feature Importances')
plt.show()


# Tune hyperparameters: max_depth, min_samples_leaf
model = RandomForestClassifier(random_state=0)
cv = KFold(n_splits=10, shuffle=True, random_state=0)
params = {'max_depth': np.arange(2, 15), 'min_samples_leaf': np.arange(1, 6)}
cv_model = GridSearchCV(model, param_grid=params, scoring='accuracy', refit=True, return_train_score=True, cv=cv)
cv_model.fit(x_train, y_train)
cv_results = pd.DataFrame(cv_model.cv_results_)

rfc_best = cv_model.best_estimator_
rfc_best.fit(x_train, y_train)
yhat_train = rfc_best.predict(x_train)
yhat_test = rfc_best.predict(x_test)

print('Best maximum depth: %i' % cv_model.best_params_['max_depth'])
print('Best minimum samples leaf: %i' % cv_model.best_params_['min_samples_leaf'])
print('CV accuracy: %.3f' % cv_model.best_score_)

print('Accuracy (train): %.3f' % accuracy_score(y_train, yhat_train))
print('Accuracy (test): %.3f' % accuracy_score(y_test, yhat_test))
print('Precision (train): %.3f' % precision_score(y_test, yhat_test))
print('Precision (test): %.3f' % precision_score(y_test, yhat_test))
print('Recall (train): %.3f' % recall_score(y_test, yhat_test))
print('Recall (test): %.3f' % recall_score(y_test, yhat_test))
print('AUC: %.3f' % roc_auc_score(y_test, yhat_test))
```
